## Matrix Invers

Dalam aljabar linear matrix persegi A n x n disebut invertible (atau juga disebut nonsingular, nondegenerate, atau biasanya digunakan secara umum).

Jika terdapat sebuah matrix persegi B n x n dan **AB = BA = I**, dimana I adalah dinotasikan dengan matrix Identitas n x n dan perkalian yang digunakan adalah perkalian matrix biasa. Maka Matrix B secara unik ditentukan oleh Matrix A, dan disebut Inverse dari Matrix A. dinotasikan dengan $\begin{align*}A^-1\end{align*}$. Matrix inversion adalah proses untuk menemukan invers marix dari invertible Matrix. Lain halnya Matrix persegi yang tidak dapat dibalik disebut Matrix singular atau matrix tunggal atau degenerate.

$
\begin{align*}

\end{bmatrix}
\end{align*}
$

Misal Matrix $
\begin{align*}
B =
\begin{bmatrix}
-1&3\\

\end{bmatrix}
\end{align*}
$
Matrix AB = BA = I, Maka Matrix B adalah Matrix $\begin{align*}A^-1\end{align*}$.

Misal kita cari nilai B nya dulu yaitu invers dari matrix A, Bisa kita buktikan dengan code,

Matrix I disini adalah Matrix Identitasnya=

$$

````python
# contoh Matrix A dan B 2 x 2
import numpy as np

A = np.array([[2,3],[1,1]])
B = np.array([[-1,3],[1,-2]])

print(f"A = \n{A}")
print(f"B = \n{B}")
print("\n")

def RowSwap(A,k,l):
# =============================================================================
#     A is a NumPy array.  RowSwap will return duplicate array with rows
#     k and l swapped. (menukarkan persamaan)
# =============================================================================
    m = A.shape[0]  # m is number of rows in A (jumlah baris A)
    n = A.shape[1]  # n is number of columns in A (jumlah kolom A)

    B = np.copy(A).astype('float64')

    for j in range(n):
        temp = B[k][j]
        B[k][j] = B[l][j]
        B[l][j] = temp

    return B

def RowScale(A,k,scale):
# =============================================================================
#     A is a NumPy array.  RowScale will return duplicate array with the
#     entries of row k multiplied by scale. (mengalikan entry baris A dengan skala)
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[k][j] *= scale

    return B

def RowAdd(A,k,l,scale):
# =============================================================================
#     A is a numpy array.  RowAdd will return duplicate array with row
#     l modified.  The new values will be the old values of row l added to
#     the values of row k, multiplied by scale. (menambahkan nilai baris k dikalikan dengan skala)
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[l][j] += B[k][j]*scale

    return B

print(f"Matrix AB =  Matrix BA \n{A@B} =\n \n{B@A}")
print("\n")

````

## Sifat sifat Matrix Invers

Matrix invers adalah sifat yang berlaku pada matrix yang merupakan kebalikan dari Matrix asal. Matrix invers dapat ditulis dengan simbol $\begin{align*}A^-1\end{align*}$ yang berarti jika dikalikan dengan Matrix A akan menghasilkan Matrix Identitas atau Matrix $\begin{align*}I\end{align*}$. Matrix Identitas adalah Matrix yang memiliki angka 1 sebagai diagonal utamanya dan selain itu angka 0 nilainya.

**Berikut adalah sifat sifat Matrix Invers:**
1. Matrix Invers hanya dimiliki oleh Matrix persegi atau Matrix yang memiliki baris dan kolom yang sama
2. Matrix invers hanya ada pada Matrix yang non-singular, yaitu Determinannya tidak sama dengan 0
3. Invers Matrix bersifat tunggal, artinya tidak ada matrix invers selain yang menjadi invers dari matrix tersebut. Jika B dan C adalah invers dari Matrix A, maka B = C.
4. Invers dari invers suatu Matrix adalah Matrix itu sendiri
5.  Invers dari hasil kali dua atau lebih matriks adalah hasil kali invers-inversnya dalam urutan terbalik. Jika A dan B keduanya nonsingular, maka  $\begin{align*}(AB)^-1 = B^-1 A^-1\end{align*}$.
6. **Ax = b** maka **x = $\begin{align*}A^-1\end{align*}$b** , x adalah variabel dari koefisien A, dan b adalah konstanta
7. Matrix Invers A dikali dengan A maka sama dengan Matrix Identitas. $\begin{align*}A^-1\end{align*}$ x $\begin{align*}A \end{align*}$ = $\begin{align*}I\end{align*}$


## Mencari invers dengan eliminasi gaus jordan

Dalam penerapan gauss Jordan dengan menggunakan Operasi Baris Elementer atau disebut OBE. Operai OBE terdiri dari 3 cara, yaitu:
1. Menambahkan kelipatan baris ke baris yang lain (RowAdd)
2. mengkalikan baris dengan bilangan selain nol (RowScale)
3. Menukarkan posisi baris satu dengan baris yang lain (RowSwap)

Untuk mencari Invers dari sebuah Matrix kita harus memperhatikan persamaan - persamaannya. Untuk bisa memasukkannya dalam bentuk matrix kita pisah antara koefisien yang ada disebelah ruas kiri dan konstanta yang ada disebelah ruas kanan.

````python
def RowSwap(A,k,l):
# =============================================================================
#     A is a NumPy array.  RowSwap will return duplicate array with rows
#     k and l swapped. (menukarkan persamaan)
# =============================================================================
    m = A.shape[0]  # m is number of rows in A (jumlah baris A)
    n = A.shape[1]  # n is number of columns in A (jumlah kolom A)

    B = np.copy(A).astype('float64')

    for j in range(n):
        temp = B[k][j]
        B[k][j] = B[l][j]
        B[l][j] = temp

    return B

def RowScale(A,k,scale):
# =============================================================================
#     A is a NumPy array.  RowScale will return duplicate array with the
#     entries of row k multiplied by scale. (mengalikan entry baris A dengan skala)
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[k][j] *= scale

    return B

def RowAdd(A,k,l,scale):
# =============================================================================
#     A is a numpy array.  RowAdd will return duplicate array with row
#     l modified.  The new values will be the old values of row l added to
#     the values of row k, multiplied by scale. (menambahkan nilai baris k dikalikan dengan skala)
# =============================================================================
    m = A.shape[0]  # m is number of rows in A
    n = A.shape[1]  # n is number of columns in A

    B = np.copy(A).astype('float64')

    for j in range(n):
        B[l][j] += B[k][j]*scale

    return B
````
