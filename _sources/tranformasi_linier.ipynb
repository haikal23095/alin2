## Tranformasi Linier
Transformasi linear merupakan dasar dalam aljabar linear yang
berbentuk fungsi. Transformasi linear yang dimaksud adalah perpindahan
dari satu ruang yang biasa dinamakan dengan domain ke ruang lain yang
dinamakan kodomain. Salah satu pembahasan dalam perkuliahan aljabar
adalah mengenai transformasi linear yaitu suatu fungsi yang dapat
memetakan suatu ruang vektor ke ruang vektor yang lain, sehingga operasi
standar pada ruang vektor (penjumlahan dan perkalian dengan skalar) tetap
berlaku.

image.png

**Cara Menentukan Tranformasi Linier:**
1. Pusat titik tidak berubah
2. Garis lurus ditetapkan sebagai garis lurus

**Atau secara umum pemetaan T adalah tranformasi Linier jika**
1. $T(u + v) = T(u) + (v)$
2. $T(\alpha v)= \alpha  T(v)$

## Transformasi Linier dalam Matriks
Transformasi linier adalah pemetaan antara dua ruang vektor yang memiliki sifat-sifat tertentu. Dalam konteks matriks, transformasi linier dapat direpresentasikan sebagai perkalian matriks.

Sifat-sifat transformasi linier:

* Memelihara asal: Transformasi linier memetakan vektor nol ke vektor nol.
* Memelihara kombinasi linear: Transformasi linier memetakan kombinasi linear dari vektor-vektor ke kombinasi linear dari bayangan vektor-vektor tersebut.
* Representasi matriks:

Misalkan T adalah transformasi linier yang memetakan ruang vektor V ke ruang vektor W. Matriks representasi T, dilambangkan dengan [T], adalah matriks yang diperoleh dengan mengalikan setiap vektor di V dengan kolom-kolom matriks, dan menempatkan hasilnya sebagai baris-baris matriks baru.

Contoh transformasi linier dalam matriks:

* Rotasi: Memutar vektor di sekitar sumbu tertentu. Matriks rotasi memiliki elemen yang ditentukan oleh sudut rotasi dan sumbu rotasi.
* Refleksi: Memantulkan vektor di sepanjang garis atau bidang tertentu.
* Matriks refleksi memiliki elemen yang ditentukan oleh normal garis atau bidang refleksi.
* Proyeksi: Memproyeksikan vektor ke subruang tertentu. Matriks proyeksi memiliki elemen yang ditentukan oleh basis subruang.
* Penskalaan: Mengubah skala vektor dengan faktor tertentu. Matriks penskalaan memiliki elemen yang ditentukan oleh faktor skala.

Manfaat transformasi linier dalam matriks:

* Memvisualisasikan transformasi linier secara geometris.
* Menghitung efek transformasi linier pada vektor-vektor tertentu.
* Menyelesaikan sistem persamaan linier yang melibatkan transformasi linier.
* Menganalisis sifat-sifat transformasi linier.

##Tranformasi Linier dengan Refleksi
Refleksi atau pencerminan pada tranformasi linear adalah adalah salah satu metode tranformasi linear yang digunakan untuk mencerminkan suatu Node dengan fungsi matrix. Berikut adalah sifat-sifat Refleksi:
1. Objek yang direfleksikan tidak mengalami perubahan benntuk dan ukuran
2. Jarak objek dari cermin (sumbu X/Y) adalah sama dengan jarak bayangan dengan cermin tersbut.

* Untuk bisa dicerminkan ke sumbu X fungsi matrix harus seperti berikut:
$
\begin{align*}
f(Matrix) =
\begin{bmatrix}
1 & 0\\
0 & -1\\
\end{bmatrix}
\end{align*}
$

* Untuk bisa dicerminkan ke sumbu Y fungsi matrix harus seperti berikut:
$
\begin{align*}
f(Matrix) =
\begin{bmatrix}
-1 & 0\\
0 & 1\\
\end{bmatrix}
\end{align*}
$

```python
# import

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
print(f"Versi Matplotlib = {matplotlib.__version__}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat matriks asli
A = np.array([[1, 2], [3, 4], [5, 6]])

# Membuat matriks transformasi untuk refleksi terhadap sumbu x
T = np.array([[1, 0], [0, -1]])

# Melakukan transformasi linear dengan metode refleksi
A_transformed = np.dot(A, T)

# Menggambar plot
plt.figure(figsize=(6, 6))

# Plot matriks asli
plt.plot(A[:, 0], A[:, 1], 'o-', label='Original')

# Plot matriks yang telah ditransformasi
plt.plot(A_transformed[:, 0], A_transformed[:, 1], 'o-', label='Transformed')

# Menambahkan garis sumbu x dan y
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```

## Tranformasi Linier Rotasi
Dalam aljabar linear, matriks rotasi adalah matriks transformasi yang digunakan untuk melakukan rotasi dalam ruang Euclidean 2 dimensi. Misalnya, dengan menggunakan konvensi di bawah ini, matriks
$
\begin{align*}
R =
\begin{bmatrix}
cos\theta &-sin\theta \\
sin\theta & cos\theta\\
\end{bmatrix}
\end{align*}
$
memutar titik-titik pada bidang xy berlawanan arah jarum jam melalui θ terhadap sumbu x terhadap titik asal sistem koordinat kartesius dua dimensi. Untuk melakukan rotasi pada titik bidang dengan koordinat standar v = (x, y), harus ditulis sebagai vektor kolom, dan dikalikan dengan matriks R:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan array a (vektor yang akan dirotasi)
a = np.array([[2, 6]])

# Definisikan sudut rotasi (dalam derajat) dan konversi ke radian
theta = 90  # Sudut rotasi dalam derajat (90 derajat)
theta_rad = np.radians(theta)  # Konversi ke radian

# Definisikan matriks rotasi
R = np.array([[np.cos(theta_rad), -np.sin(theta_rad)],
              [np.sin(theta_rad), np.cos(theta_rad)]])

# Hitung hasil rotasi
c = a @ R
print("Hasil rotasi matriks a dengan sudut", theta, "derajat adalah:", c)
# Output: Hasil rotasi matriks a dengan sudut 45 derajat adalah: [[0. 5.65685425]]

# Visualisasi transformasi vektor sebelum dan sesudah rotasi
plt.figure(figsize=(12, 6))

# Vektor asli
plt.quiver(0, 0, a[0, 0], a[0, 1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vektor asli (a)')

# Vektor setelah rotasi
plt.quiver(0, 0, c[0, 0], c[0, 1], angles='xy', scale_units='xy', scale=1, color='red', label='Vektor rotasi (c)')

# Pengaturan plot
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('label x')
plt.ylabel('label y')
plt.title('Visualisasi Rotasi Vektor oleh Matriks')

# Refleksi visualisasi:
# Vektor asli (a) = [4, 4] ditampilkan dalam warna biru.
# Vektor setelah rotasi (c) = hasil rotasi a sebesar 45 derajat ditampilkan dalam warna merah.
# Rotasi ini mengubah arah vektor a sesuai dengan sudut yang diberikan.

# Menampilkan grafik
plt.show()
```

## Tranformasi Linier Scalling

Transformasi linear dengan metode scaling pada matriks adalah proses mengubah matriks asli dengan cara mengalikan setiap elemen matriks tersebut dengan suatu bilangan skalar. Dalam konteks grafis komputer, scaling adalah proses mengubah ukuran objek. Dalam konteks visual, metode scaling akan mengubah ukuran objek tanpa mengubah bentuk aslinya. Misalnya, jika kita memiliki titik pada koordinat (2, 3) dan kita melakukan scaling 2x pada sumbu x dan 3x pada sumbu y, maka titik tersebut akan berpindah ke koordinat (4, 9).

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat matriks asli
A = np.array([[1, 2], [3, 4], [5, 6]])

# Membuat matriks transformasi untuk scaling
# Misalnya kita ingin melakukan scaling 2x pada sumbu x dan 3x pada sumbu y
T = np.array([[2, 0], [0, 3]])

# Melakukan transformasi linear dengan metode scaling
A_transformed = np.dot(A, T)

# Menggambar plot
plt.figure(figsize=(6, 6))

# Plot matriks asli
plt.plot(A[:, 0], A[:, 1], 'o-', label='Original')

# Plot matriks yang telah ditransformasi
plt.plot(A_transformed[:, 0], A_transformed[:, 1], 'o-', label='Transformed')

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```
