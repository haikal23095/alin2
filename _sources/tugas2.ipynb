### Pengertian Matrix:

### Operasi Matrix
Matrix juga bisa kita lakukan operasi yaitu Penjumlahan, Perkalian, dan Pengurangan. Untuk penjumlahan dan pengurangan hanya bisa dilakukan oleh matrix yang memiliki ordo yang sama. Sedangkan perkalian antar matrix bisa dilakukan jika baris pada matrix_a sama dengan jumlah kolom pada matrix_b.

#### Operasi Penjumlahan Matrix
Operasi penjumlahan matrix adalah penjumlahan antara nilai nilai yang ada pada dua buah Matrix. Penjumlahan dapat dioperasikan secara langsung antara baris dan kolom yang sama.

image.png

Seperti contoh berikut yang diimplementasikan bentuk dalam code berikut:

```python
# Matrix A+B
print(f"Matrix A : \n{A}")
print(f"Matrix B : \n{B}")
E = (A+B)
print(f"Matrix E = Matrix A + Matrix B:\n{E}")
print(f"Ukuran Matrix E = {E.shape}")
```

#### Operasi Perkalian Matrix
Operasi perkalian matrix ada dua macam
- perkalian Matrix dengan sebuah nilai tertentu. Pada perkalian ini kita bisa operasikan langsung antara nilai nilai yang ada di dalam matrix dengan sebuah nilai tertentu.
- perkalian antara dua buah Matrix. perkalian antara dua buah matrix dapa disimbolkan dengan matrix_a @ matrix_b atau matrix_a . matrix_b. Syarat untuk mengkalikan antara dua buah matrix yaitu jumlah kolom pada matrix_a harus sama dengan jumlah baris pada matrix_b.

image.png

```python
# Perkalian Matrix A dengan angka 2 yang menghasilkan matrix F
print(f"Matrix A = \n{A}")
F = 2*(A)
print(f"Perkalian 2 * Matrix A =\n{F}")
print(f"Ukuran Matrix = {F.shape}")
```

```python
# Matrix A x C
print(f"Matrix A 3x3: \n{A}")
print(f"Matrix C 3x2: \n{C}")
G = A @ C
print(f"Matrix G = Matrix A @ Matrix C:\n{G}")
print(f"Ukuran Matrix = {G.shape}")
```

### Cek Persamaan
Matrix juga berlaku sifat Asosiatif seperti berikut:

Matrix_A(Matrix_B x Matrix_C) = (Matrix_A x Matrix_B)Matrix_C

```python
print(f"Matrix_A =\n{A}")
print(f"Matrix_B =\n{B}")
print()

print(f"Marix_A(Matrix_B x Matrix_C) = (Matrix_A x Matrix_B)Matrix_C\n")

Matrix1 = A @ (B @ C)
Matrix2 = (A @ B) @ C
Matrix1 == Matrix2
```

Bisa juga seperti ini:

Matrix c(AB) = Matrix (cA)B = Matrix A(cB)

```python
c = 2
Matrix_3 = c * (A @ B)
Matrix_4 = (c * A) @ B
Matrix_5 = A @ (c * B)
print(Matrix_3)
print()
print(Matrix_4)
print()
print(Matrix_5)
```

### Tranpose Matrix
Anggap Baris = m, Kolom = n

=> C = A_trans @ A

=> (n x m) @ (m x n) = n * n

contoh: [2 x 3] @ [3 x 2] = [2 x 2]

=> C = A @ A_trans

=> (m x n) @ (n x m) = m x m

contoh: [3 x 2] @ [2 x 3] = [3 x 3]



```python
# Contoh dengan Martrix 3x3

A_tran = A.transpose()
print('Matrik A sebelum di transpose=\n', A)
print()
print('Matrik A setelah di transpose=\n',A_tran)
print('Ordo A_transpose adalah',A_tran.shape)
print()
Z = A_tran @ A
print('Z = A_transpose x A =\n',Z)
print('Ordo Z adalah',Z.shape)
print()
P=A @ A_tran
print('P = A_transpose x A =\n',P)
print('Ordo P adalah',P.shape)
```

```python
# Soal dengan Matrix 3x2

Matrix_A = np.array([[-1,2,3],[3,2,1]])

Matrix_A_Trans = Matrix_A.transpose()
print('Matrix_A sebelum di transpose=\n', Matrix_A)
print()

print('Matrix A setelah di transpose=\n',Matrix_A_Trans)
print('Ordo Matrix_A_Transpose adalah',Matrix_A_Trans.shape)
print()

Matrix_Z = Matrix_A_Trans @ Matrix_A
print('Matrix_Z = Matrix_A_Transpose x Matrix_A =\n',Matrix_Z)
print('Ordo Z adalah',Z.shape)
print()

Matrix_P = Matrix_A @ Matrix_A_Trans
print('P = A_transpose x A =\n',Matrix_P)
print('ordo Matrix P adalah',Matrix_P.shape)
```

```python
Matrix_1 = [[1,3,9],
            [3,-2,6],
            [9,4,1]]

Matrix_2 = [[1,7,5],
            [2,4,5],
            [1,-1,4]]

print(f" Matrix_1 = {Matrix_1}")
print()
print(f" Matrix_2 = {Matrix_2}")

def multiply_matrix(matrix1,matrix2):
  result = [[0,0,0],[0,0,0],[0,0,0]] # Menginisialisasikan dulu matrix dengan semua value 0

  for i in range(len(Matrix_1)):
    for j in range(len(Matrix_2[0])):
      for k in range(len(Matrix_2)):
        result[i][j] += Matrix_1[i][k] * Matrix_2[k][j]
  return result

hasil = multiply_matrix(Matrix_1,Matrix_2)
print()
print(f"Hasil perkalian Matrix_1 dan Matrix 2 = {hasil}")
print()
print(f"Pembuktian untuk Matrix A @ B =\n{A @ B}")

```
