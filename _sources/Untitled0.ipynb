# Cofactor Expansion

```python
import numpy as np
```

```python
A = np.array([[3,1,3],[2,1,2],[4,2,1]])

A_invers = np.linalg.inv(A)
print(A_invers)
```

```python
# pick baris 1

A11 = np.array([[1,2],[2,1]])
A12 = np.array([[2,2],[4,1]])
A13 = np.array([[2,1],[4,2]])
det_A11 = np.linalg.det(A11)
det_A12 = np.linalg.det(A12)
det_A13 = np.linalg.det(A13)
print(det_A11)
print(det_A12)
print(det_A13)

a11 = A[0][0]
a12 = A[0][1]
a13 = A[0][2]

print(a11)
print(a12)
print(a13)

c11 = (-1)**(1+1) * det_A11
c12 = (-1)**(1+2) * det_A12
c13 = (-1)**(1+3) * det_A13

print(c11)
print(c12)
print(c13)

row1 = a11 * c11 + a12 * c12 + a13 * c13

print(row1)
```
