```python
import numpy as np
```

```python
# buat matrix A (5x5)

A = np.array([[4,2,1,1,3],[3,2,1,2,2],[4,1,1,1,1],[1,3,4,1,2],[2,2,1,1,3]])
print(f'matrix A  = \n{A}')
```

```python
# mencari invers matrix dengan fungsi numpy np.linalg.inv
print(f'A invers = \n{np.linalg.inv(A).round(2)}')
```

```python
B = A @ np.linalg.inv(A)
print(B.round(1))
```

```python
# A_inv = 1/det_A * adj A

A_det = np.linalg.det(A)

print(A_det)

```
